```python
import pandas as pd
from sklearn import linear_model
import tkinter as tk
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

TeslaData = pd.read_csv("TESLA.csv")

X = TeslaData[['Open','High', 'Low']]
Y = TeslaData['Adj Close']

RegModel = linear_model.LinearRegression()
RegModel.fit(X, Y)

root = tk.Tk()
root.title('TESLA Stock Prediction Model')
canvas = tk.Canvas(root)

canvas1 = tk.Canvas(root, width = 500, height = 300)
canvas1.pack()

label1 = tk.Label(root, text='Enter Open Price: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root)
canvas1.create_window(270, 100, window=entry1)

label2 = tk.Label(root, text='Enter High Price: ')
canvas1.create_window(100, 120, window=label2)

entry2 = tk.Entry (root)
canvas1.create_window(270, 120, window=entry2)

label3 = tk.Label(root, text='Enter Low Price: ')
canvas1.create_window(100, 140, window=label3)

entry3 = tk.Entry (root)
canvas1.create_window(270, 140, window=entry3)

def values():
    global Open
    Open = float(entry1.get())

    global High
    High = float(entry2.get())

    global Low
    Low = float(entry3.get())

    Prediction_result  = ('Predicted Tesla Stock: ', RegModel.predict([[Open , High, Low]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(280, 280, window=label_Prediction)

button1 = tk.Button (root, text='Predict Stock Price',command=values, bg='white')
canvas1.create_window(270, 180, window=button1)

figure3 = plt.Figure(figsize=(5,4), dpi=100)
ax3 = figure3.add_subplot(111)
ax3.scatter(TeslaData['Open'].astype(float),TeslaData['Adj Close'].astype(float), color = 'g')
scatter3 = FigureCanvasTkAgg(figure3, root)
scatter3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend(['Adj Close'])
ax3.set_xlabel('Open')
ax3.set_title('Open Vs. Adj Close')

figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(TeslaData['High'].astype(float),TeslaData['Adj Close'].astype(float), color = 'g')
scatter4 = FigureCanvasTkAgg(figure4, root)
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend(['Adj Close'])
ax4.set_xlabel('High')
ax4.set_title('High Vs. Adj Close')

figure5 = plt.Figure(figsize=(5,4), dpi=100)
ax5 = figure5.add_subplot(111)
ax5.scatter(TeslaData['Low'].astype(float),TeslaData['Adj Close'].astype(float), color = 'g')
scatter5 = FigureCanvasTkAgg(figure5, root)
scatter5.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax5.legend(['Adj Close'])
ax5.set_xlabel('Low')
ax5.set_title('Low Vs. Adj Close')

lblTitle = tk.Label(root, text = 'Predictions')

root.mainloop()
```
